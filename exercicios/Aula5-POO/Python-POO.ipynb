#### Exemplo - A classe Time

```python
"""Class Time with read-write properties."""

class Time:
    """Class Time with read-write properties."""

    def __init__(self, hour=0, minute=0, second=0):
        """Initialize each attribute."""
        self.hour = hour  # 0-23
        self.minute = minute  # 0-59
        self.second = second  # 0-59

    @property
    def hour(self):
        """Return the hour."""
        return self._hour

    @hour.setter
    def hour(self, hour):
        """Set the hour."""
        if not (0 <= hour < 24):
            raise ValueError(f'Hour ({hour}) must be 0-23')

        self._hour = hour

    @property
    def minute(self):
        """Return the minute."""
        return self._minute

    @minute.setter
    def minute(self, minute):
        """Set the minute."""
        if not (0 <= minute < 60):
            raise ValueError(f'Minute ({minute}) must be 0-59')

        self._minute = minute

    @property
    def second(self):
        """Return the second."""
        return self._second

    @second.setter
    def second(self, second):
        """Set the second."""
        if not (0 <= second < 60):
            raise ValueError(f'Second ({second}) must be 0-59')

        self._second = second

    def set_time(self, hour=0, minute=0, second=0):
        """Set values of hour, minute, and second."""
        self.hour = hour
        self.minute = minute
        self.second = second

    def __repr__(self):
        """Return Time string for repr()."""
        return (f'Time(hour={self.hour}, minute={self.minute}, ' + 
                f'second={self.second})')

    def __str__(self):
        """Return Time string in 12-hour clock format."""
        return (('12' if self.hour in (0, 12) else str(self.hour % 12)) + 
                f':{self.minute:0>2}:{self.second:0>2}' + 
                (' AM' if self.hour < 12 else ' PM'))
```

### Simulando atributos privados
Em linguagens de programao como C++, Java e C#, as classes declaram explicitamente qual classe os membros so acessveis ao pblico. 
- Membros da classe que no podem ser acessados fora da classe so definidos como **private**  e so visveis apenas dentro da classe que os define. 

Programadores Python costumam usar atributos "privados" para dados ou mtodos de utilidade que so essenciais para a funcionamento interno, mas no fazem parte da interface pblica da classe.

Como visto, os atributos dos objetos Python esto sempre acessveis. No entanto, Python tem
uma conveno de nomenclatura para atributos privados. Suponha que queremos criar um objeto da classe Time e evitar a seguinte declarao de atribuio:

Este comando definiria a hora com um valor invlido. Em vez de **\_hour**, podemos nomear o atributo **\_\_hour** com dois sublinhados  esquerda. Esta conveno indica que **\_\_hour**  "privado" e no deve ser acessvel aos clientes da classe. 

Para ajudar a evitar que os clientes acessem atributos "privados", o Python os renomeia precedendo o nome do atributo com **\_ClassName**, como em
**\_Time\_\_hours**. Isso  chamado de mutilao de nome (name mangling). Se voc tentar atribuir a **\_\_hour**, como em:

Python gera um **AttributeError**, indicando que a classe no tem um atributo **\_\_hour**.

## Herana

### Polimorfismo

Habilidade de objetos de classe diferentes, relacionados por herana, de responder de forma diferente  mesma chamada.

Considere uma hierarquia contendo diferentes tipos de funcionrios no aplicativo de folha de pagamento de uma empresa para discutir o relacionamento entre uma classe base e sua subclasse. 
- Todos os funcionrios da empresa tm muito em comum,

Primeiro, apresentamos a classe base **CommissionEmployee**. Em seguida, criamos uma subclasse **SalariedCommissionEmployee** que herda da classe **CommissionEmployee**. 

A classe base **CommissionEmployee** fornece os seguintes recursos:
- Mtodo **\_\_init\_\_**, que cria os atributos de dados **\_first\_name**,**\_last\_name** e **\_ssn** (nmero da Previdncia Social), e usa os configuradores de propriedades **gross_sales** e **commission_rate** para criar seus atributos de dados correspondentes.
- Propriedades _somente leitura_ **first\_name**, **last\_name** e
**ssn**, que retorna os atributos de dados correspondentes.
- Propriedades de _leitura e gravao_ **gross\_sales** e **comission\_rate**, em que os **setters** executam a validao de dados.
 Mtodo **earning**, que calcula e retorna os ganhos de um **CommissionEmployee**
 Mtodo **\_\_repr\_\_** que retorna uma representao (string) de **CommissionEmployee**.

```python
# commmissionemployee.py
"""CommissionEmployee base class."""
from decimal import Decimal

class CommissionEmployee:
    """An employee who gets paid commission based on gross sales."""

    def __init__(self, first_name, last_name, ssn, 
                 gross_sales, commission_rate):
        """Initialize CommissionEmployee's attributes."""
        self._first_name = first_name
        self._last_name = last_name
        self._ssn = ssn
        self.gross_sales = gross_sales  # validate via property
        self.commission_rate = commission_rate  # validate via property

    @property
    def first_name(self):
        return self._first_name

    @property
    def last_name(self):
        return self._last_name

    @property
    def ssn(self):
        return self._ssn

    @property
    def gross_sales(self):
        return self._gross_sales

    @gross_sales.setter
    def gross_sales(self, sales):
        """Set gross sales or raise ValueError if invalid."""
        if sales < Decimal('0.00'):
            raise ValueError('Gross sales must be >= to 0')
        
        self._gross_sales = sales
        
    @property
    def commission_rate(self):
        return self._commission_rate

    @commission_rate.setter
    def commission_rate(self, rate):
        """Set commission rate or raise ValueError if invalid."""
        if not (Decimal('0.0') < rate < Decimal('1.0')):
            raise ValueError(
               'Interest rate must be greater than 0 and less than 1')
        
        self._commission_rate = rate

    def earnings(self):
        """Calculate earnings."""   
        return self.gross_sales * self.commission_rate

    def __repr__(self):
        """Return string representation for repr()."""
        return ('CommissionEmployee: ' + 
            f'{self.first_name} {self.last_name}\n' +
            f'social security number: {self.ssn}\n' +
            f'gross sales: {self.gross_sales:.2f}\n' +
            f'commission rate: {self.commission_rate:.2f}')
```

#### Declarando a subclasse SalariedCommissionEmployee

A subclasse **SalariedCommissionEmployee** herda a maior parte de seus recursos da classe **CommissionEmployee**.

Um **SalariedCommissionEmployee**  um **CommissionEmployee** (porque a herana transmite os recursos da classe **CommissionEmployee**), mas a classe **SalariedCommissionEmployee** tambm tem os seguintes recursos:
- Mtodo \_\_init\_\_ inicializa todos os dados herdados da classe **CommissionEmployee**, ento usa o **setter** da propriedade **base_salary** para criar um atributo de dados **\_base\_salary**.
- Propriedade de leitura e gravao **base_salary**, em que o **setter** executa a validao dos dados.
- Uma verso personalizada dos mtodo **earnings**.
- Uma verso personalizada do mtodo **\_\_repr\_\_**.

```python
# salariedcommissionemployee.py
"""SalariedCommissionEmployee derived from CommissionEmployee."""
from decimal import Decimal

class SalariedCommissionEmployee(CommissionEmployee):
    """An employee who gets paid a salary plus 
    commission based on gross sales."""

    def __init__(self, first_name, last_name, ssn, 
                 gross_sales, commission_rate, base_salary):
        """Initialize SalariedCommissionEmployee's attributes."""
        super().__init__(first_name, last_name, ssn, 
                         gross_sales, commission_rate)
        self.base_salary = base_salary  # validate via property

    @property
    def base_salary(self):
        return self._base_salary

    @base_salary.setter
    def base_salary(self, salary):
        """Set base salary or raise ValueError if invalid."""
        if salary < Decimal('0.00'):
            raise ValueError('Base salary must be >= to 0')
        
        self._base_salary = salary

    def earnings(self):
        """Calculate earnings."""   
        return super().earnings() + self.base_salary

    def __repr__(self):
        """Return string representation for repr()."""
        return ('Salaried' + super().__repr__() +      
            f'\nbase salary: {self.base_salary:.2f}')
```

Especifica que a classe **SalariedCommissionEmployee** herda de **CommissionEmployee**.

- Embora voc no veja os atributos de dados, propriedades e mtodos da classe **CommissionEmployee** na classe **SalariedCommissionEmployee**, eles fazem parte da nova classe.

### O mtodo \_\_init\_\_ e funo interna *super*

<font color=red>Cada subclasse **\_\_init\_\_** deve chamar explicitamente o **\_\_init\_\_** de sua classe base para inicializar os atributos de dados herdado da classe base.</font> 

- Esta chamada deve ser a primeira instruo no mtodo **\_\_init\_\_** da subclasse. 

O mtodo **\_\_init\_\_** de **SalariedCommissionEmployee** chama explicitamente o mtodo **\_\_init\_\_** da classe **CommissionEmployee** para inicializar a parte da classe base
de um objeto SalariedCommissionEmployee (ou seja, os cinco atributos de dados herdados de
classe CommissionEmployee). 

**<center>super().\_\_init\_\_(first_name, last_name, ssn, gross_sales, commission_rate) </center>**

A notao **super().\_\_init\_\_** usa a funo embutida *super* para localizar e chamar o mtodo **\_\_init\_\_** da classe base, passando os cinco argumentos que
inicializam os atributos de dados herdados.

### Substituindo o mtodo _earnings_

Para calcular o ganhos de um **SalariedCommissionEmployee**, o mtodo **earnings** da classe **SalariedCommissionEmployee** substitui o mtodo de mesmo nome da classe **CommissionEmployee** .

Portanto, ao processar um objeto em tempo de execuo, seu tipo no importa. Contanto que o

## Estudo de caso: embaralhando e dando as cartas

Este estudo de caso apresenta duas classes personalizadas que voc pode usar para embaralhar e distribuir um baralho de cartas. 

- A classe **Card** representa uma carta de jogo que tem uma face ('s', '2', '3', ..., 'Valete', 'Dama', 'Rei') e um naipe ('Copas', 'Ouros', 'Paus', 'Espadas'). 
- A classe **DeckOfCards** representa um baralho de 52 cartas como uma lista de objetos de cartas (**Card**). 

```python
# card.py
"""Card class that represents a playing card and its image file name."""
class Card:
    FACES = ['Ace', '2', '3', '4', '5', '6',
             '7', '8', '9', '10', 'Jack', 'Queen', 'King']
    SUITS = ['Hearts', 'Diamonds', 'Clubs', 'Spades']

    def __init__(self, face, suit):
        """Initialize a Card with a face and suit."""
        self._face = face
        self._suit = suit

    @property
    def face(self):
        """Return the Card's self._face value."""
        return self._face

    @property
    def suit(self):
        """Return the Card's self._suit value."""
        return self._suit

    @property
    def image_name(self):
        """Return the Card's image file name."""
        return str(self).replace(' ', '_') + '.png'

    def __repr__(self):
        """Return string representation for repr()."""
        return f"Card(face='{self.face}', suit='{self.suit}')"     

    def __str__(self):
        """Return string representation for str()."""
        return f'{self.face} of {self.suit}'

    def __format__(self, format):
        """Return formatted string representation."""
        return f'{str(self):{format}}'

```

```python
# deck.py
"""Deck class represents a deck of Cards."""
import random 
#import Card

class DeckOfCards:
    NUMBER_OF_CARDS = 52  # constant number of Cards

    def __init__(self):
        """Initialize the deck."""
        self._current_card = 0
        self._deck = []

        for count in range(DeckOfCards.NUMBER_OF_CARDS):  
            self._deck.append(Card(Card.FACES[count % 13], 
                Card.SUITS[count // 13]))

    def shuffle(self):
        """Shuffle deck."""
        self._current_card = 0
        random.shuffle(self._deck)    

    def deal_card(self):
        """Return one Card."""
        try:
            card = self._deck[self._current_card]
            self._current_card += 1
            return card
        except:
            return None  

    def __str__(self):
        """Return a string representation of the current _deck."""
        s = ''

        for index, card in enumerate(self._deck):
            s += f'{self._deck[index]:<19}'
            if (index + 1) % 4 == 0:
                s += '\n'
        
        return s

```

## Uma breve introduo s novas _Data Classes_ do Python 3.7

Muito embora as tuplas nomeadas permitam que voc faa referncia a seus membros pelo nome, elas ainda so apenas tuplas, no classes. 

Para alguns dos benefcios das tuplas nomeadas, alm dos recursos tradicionais que as classes Python fornecem, voc pode usar as novas [data classes](https://www.python.org/dev/peps/pep-0557/) do Python 3.7.

As classes de dados (data classes) esto entre os novos recursos mais importantes do Python 3.7. 
- Elas te ajudam a construir classes mais rpido usando uma notao mais concisa e gerando automaticamente cdigo "clich", isto , comum na maioria das classes. 
- Elas podem se tornar a forma preferida de definir muitas Classes Python.

A maioria das classes que voc definir fornece um mtodo **\_\_init\_\_** para criar e inicializar o atributos e um mtodo **\_\_repr\_\_** para especificar a representao de string personalizada de um objeto. 
- Se um classe tem muitos atributos de dados, criar esses mtodos pode ser entediante.
- As classes de dados (data classes) geram automaticamente os atributos de dados e os mtodos **\_\_init\_\_** e **\_\_repr\_\_** para voce. 
- Isso pode ser particularmente til para classes que agregam itens de dados relacionados
    - Por exemplo, em um aplicativo que processa registros CSV, voc pode querer uma classe que representa os campos de cada registro como atributos de dados em um objeto.

```python
"""Card data class with class attributes, data attributes,
autogenerated methods and explicitly defined methods."""
from dataclasses import dataclass
from typing import ClassVar, List

@dataclass
class Card:
    FACES: ClassVar[List[str]] = ['Ace', '2', '3', '4', '5', '6', '7', 
                                  '8', '9', '10', 'Jack', 'Queen', 'King']
    SUITS: ClassVar[List[str]] = ['Hearts', 'Diamonds', 'Clubs', 'Spades']

    face: str
    suit: str

    @property
    def image_name(self):
        """Return the Card's image file name."""
        return str(self).replace(' ', '_') + '.png'

    def __str__(self):
        """Return string representation for str()."""
        return f'{self.face} of {self.suit}'
    
    def __format__(self, format):
        """Return formatted string representation."""
        return f'{str(self):{format}}'
```

### Importando dos mdulos _dataclasses_ e _typing_

O mdulo da biblioteca padro Python, **_dataclasses_** define [decoradores](https://docs.python.org/3/library/dataclasses.html#module-level-decoratorsclasses-and-functions) e funes para implementao de classes de dados. 

- Usamos o decorador **@dataclass** para especificar que uma nova classe  uma classe de dados e faz com que vrios cdigos sejam escritos para voc. 
- Usamos **ClassVar** e **List** do mdulo **_typing_** para indicar que **FACES** e **SUITS** so de _variveis de classe_ que se referem a listas.

### Anotaes de varivel: atributos de classe

Ao contrrio das classes regulares, as classes de dados <font color=red>declaram</font> atributos de classe e atributos de dados dentro da classe, mas fora dos mtodos da classe. 
- Em uma classe regular, apenas os atributos da classe so declarados dessa forma,
- os atributos de dados normalmente so criados em **\_\_init\_\_**. 

As classes de dados exigem informaes, ou _dicas_, para distinguir atributos de classe de atributos de dados, que tambm afeta os detalhes de implementao dos mtodos gerados automaticamente.

O cdigo abaixo inicializa os atributos de classe FACES e SUITS

**: ClassVar[List[str]]**  uma [anotao de varivel](https://www.python.org/dev/peps/pep-0526/) (s vezes chamado de _dica de tipo_) especificando que FACES  um atributo de classe (ClassVar) que se refere a uma lista de strings (List [str]). SUITS tambm  um atributo de classe que se refere a uma lista de strings.
- Variveis de classe so inicializadas em suas definies e so especficas para a classe, e no a objetos da classe. 
- Mtodos **\_\_init\_\_**, **\_\_repr\_\_** e **\_\_eq\_\_**, no entanto, so para uso com objetos da classe. Quando uma classe de dados gera esses mtodos, ela inspeciona todas as variveis anotaes e inclui apenas os atributos de dados nas implementaes do mtodo.

### Anotaes de variveis: atributos de dados
Normalmente, criamos os atributos de dados de um objeto no mtodo **\_\_init\_\_** da classe (ou mtodos chamado por **\_\_init\_\_**) por meio de atribuies do tipo **self.attribute_name = value**. 
- Como uma classe de dados gera automaticamente seu mtodo **\_\_init\_\_**, precisamos de outra maneira de especificar os atributos de dados na definio de uma classe de dados. 

Como os atributos de classe, cada atributo de dados deve ser declarado com uma anotao de varivel. As linhas abaixo definem os atributos de dados **face** e **suit**. 
- A anotao de varivel ": str" indica que cada um deve se referir a objetos string

### Definindo uma propriedade e outros mtodos

Classes de dados so classes, portanto, podem conter propriedades e mtodos bem como fazer parte de hierarquias de classes. 

Para esta classe de dados **Card**, definimos a mesma propriedade **image_name** somente leitura e mtodos especiais personalizados **\_\_str\_\_** e **\_\_format\_\_** como em nossa classe Card original.

- Mesmo com anotaes de tipo, Python ainda  uma linguagem tipada dinamicamente. Ento,  **notaes de tipo no so aplicadas no tempo de execuo**. 
    - Portanto, mesmo que a face de uma carta seja feita para ser uma string, voc pode atribuir qualquer tipo de objeto a ela.
